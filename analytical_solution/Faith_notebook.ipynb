## Function to Compute $\ket{n}$

```python
x_range = np.linspace(-5, 5, 1000)
inputs = [FunctionStateNInput(n, x_range) for n in range(4)]

# Plot the wavefunctions for the first few energy levels
plt.figure(figsize=(10, 6))

for input_data in inputs:
    psi_n = func_state_n(input_data)
    plt.plot(input_data.x, psi_n, label=f'n={input_data.n}')

plt.title('Harmonic Oscillator Wavefunctions')
plt.xlabel('x')
plt.ylabel('$\psi_n(x)$')
plt.legend()
plt.grid(True)
plt.show()
```
